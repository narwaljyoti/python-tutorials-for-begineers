```python
#Deal with categorical data sex ,smoker,day ,time
import seaborn as sns 
plt.figure(figsize=(16,9))  #increase the size of the graph 
sns.set(style="darkgrid")  #change the background color of the graph 
sns.lineplot(x="size",y="total_bill",data=tipsdata,hue="sex",style="sex",palette="Accent",dashes=True,markers=["o","<"],legend=True)
#hue parameter represent  which column in the dataframe we want to use for color encoding
#style:give style to the graph 
#palette :give the color to the graph 
#dashesTrue :will give dashes in the line graph 
#markers:market the line with any msign *,<
#legend=True means show the legend 
plt.title("Line plot",fontsize=20)
plt.xlabel("size",fontsize=30)  #resize the xlabel 
plt.show()
```

# Histogram graph /distplot :Histograms show the distribution of data by constructing bins throughout the data's range and then drawing bars to show how many observations fall into each bin.

```python
import seaborn as sns 
from scipy.stats import norms 
data=sns.load_dataset("tips")
print(data)
```

```python
  #bydefault jo bhi hum data denge usko histogram m plot krega 
sns.distplot(data['size'])   #here histogram with line plot (KDE:KERNAL DENSITY ESTIMAatION )

print("KDE is estimate the probabilty density  function continuous random variable. ")
```

```python
sns.distplot(data["total_bill"],bins=100,kde=False,rug=True,color="red",vertical=True,label="total bill")
plt.legend()
plt.show()

#bins want to print data in particular bar 
#hist=True (bydefault)  :when we want KDE LIKE only line 
#kde:want to print only hist bar 
#rug=False(bydefault):on the x -axis we got some bar , in which particular location data is available
#color="give the color of the bar "
```

```python
data['total_bill'].sort_values()  #sort the values of total_bill columns  min=3.07,max=50.81 
bins=[1,5,10,15,20,25,30,35,40,45,50,55]  #want to show in particular range  bar  

sns.distplot(data["total_bill"],bins=bins,hist_kws={'color':"green",'edgecolor':'yellow','linewidth':10})
plt.xticks(bins)
plt.show()
```

# Bargraph Barplot :relation between a categorical variable and a continuous variable.

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np 
tip=sns.load_dataset("tips")
order=["sun","thur","fri","sat"]   #set order in x-axes according to day
hue_order=['Female','Male']
#sns.barplot(x=tip['day'],y=tip['total_bill'],hue=tip['sex'],order=order)   #here find the relationship between day and total_bill
sns.barplot(x="day",y='total_bill',hue="sex",data=tip,estimator=np.max,ci=45)
#confidence interval :line in the bar will be less or more ,more value line will be less 
plt.show()
```

```python
sns.set()
kwargs={'alpha':0.9,'linestyle':'-','linewidth':5,'edgecolor':'k'}
sns.barplot(x="day",y="total_bill",hue="sex",data=tip,**kwargs)
plt.savefig("Restaurant bill in barplot")
plt.show()
```

# Heatmap:it is a graphical representation  OF 2d data .each data value represents in a matrix and has a special colour.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
x=np.linspace(1,5, 12).reshape(4,3)
print(x)  #according to value weight color density change 
```

```python
sns.heatmap(x)
plt.show()
```
